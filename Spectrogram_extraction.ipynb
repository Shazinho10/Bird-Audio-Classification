```python
def melspectrogram(save_path, bird_path, bird_name):
  #collecting all the path names from the folder
    file_list = librosa.util.find_files(bird_path, ext=['wav'])
    
    #generating mel spectrogram
    for i, j in enumerate(file_list):
        signal, sr = librosa.load(j)
        mfcc = librosa.feature.melspectrogram(signal, sr=sr, n_mels=100)
        log_mel = librosa.power_to_db(mfcc, ref=np.max)

        #save the spectrogram
        path_to_save = os.path.join(save_path, bird_name+f'spectrogram{i}')
        np.save(path_to_save, log_mel)
```
