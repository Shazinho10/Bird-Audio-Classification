```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import IPython.display as ipd
import os
```

```python
def load_files(path):
  #loading the paths
  spec_path_list = []
  for i in os.listdir(path):
    paths = os.path.join(path,i)
    spec_path_list.append(paths)
  return spec_path_list
```

```python
#training spectrograms
capuchin_train_files = load_files('capuchin_train_spectrograms')
not_capuchin_train_files = load_files('not_capuchin_train_spectrograms')
```

```python
#test spectrograms
capuchin_test_files = load_files('capuchin_test_spectrograms')
not_capuchin_test_files = load_files('not_capuchin_test_spectrograms')
```

```python
classes = {
    'capuchin' : 0,
    'not_capuchin' : 1,
}

#training json files
capuchin_json = {
    'capuchin' : capuchin_train_files,
    'not_capuchin' : not_capuchin_train_files
}


#testing json files
capuchin_test_json = {
    'capuchin' : capuchin_test_files,
    'not_capuchin' : not_capuchin_test_files
}
```

```python
X_train, y_train = [], []
for labels, training_files in capuchin_json.items():
    for audio in training_files:
        audio1 = np.load(audio)
        #reshaped_audio = np.reshape(audio, (10, 179))
        X_train.append(audio1)
        y_train.append(classes[labels])
```

```python
X_test, y_test = [], []
for labels, training_files in capuchin_test_json.items():
    for audio in training_files:
        audio2 = np.load(audio)
        #reshaped_audio = np.reshape(audio, (10, 179))
        X_test.append(audio2)
        y_test.append(classes[labels])
```

```python
def add_padding(array):
  #the array with the max size
  max_value = max([k.shape[1] for k in X_train])

  X_train_new = []
  for i in range(len(array)):
    zeros = np.zeros((array[i].shape[0], max_value - array[i].shape[1]))
    new_array = np.concatenate([array[i], zeros], axis=1)
    X_train_new.append(new_array)
  return X_train_new
```

```python
X_train_2 = add_padding(X_train)
X_test_2 = add_padding(X_test)
```

```python
X_train_arr = np.array(X_train_2)
X_test_arr = np.array(X_test_2)
```

```python
y_train_arr = np.array(y_train)
y_test_arr = np.array(y_test)
```

```python
#saving the training and testing arrays
np.save('training_testing_arrays_saved/x_train', X_train_arr)
np.save('training_testing_arrays_saved/x_test', X_test_arr)
np.save('training_testing_arrays_saved/y_train', y_train_arr)
np.save('training_testing_arrays_saved/y_test',y_test_arr)
```
